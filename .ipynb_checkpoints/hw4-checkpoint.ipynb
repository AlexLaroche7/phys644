```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import quad
```

# Question 1

```python
# Integral
integrand=lambda z: (Omega_L0+Omega_m0*(1+z)**3)**(-0.5)
a0r=lambda z: quad(integrand,0,z)[0]

# zeq
H_0=70 #km/s/Mpc
zeq=lambda Omega_m0: (2.4e4)*Omega_m0*(H_0/100)**2
```

$d_L=a_0r(1+z)$ and $d_A=\frac{a_0r}{(1+z)}$ by Mo et al. 3.101. We will plot them in units of $c/H_0$ to avoid large numbers.

```python
#compute dA and dL
plt.figure(figsize=(10,6))
zs=np.linspace(0,300,5000) # z << z_eq

# Case 2

Omega_L0=0.3
Omega_m0=0.7

a0rs=np.array([a0r(z) for z in zs])
label=r'$d_L\,(\Omega_{\Lambda,0},\Omega_{m,0})=($'+str(Omega_L0)+','+str(Omega_m0)+')'
plt.loglog(zs,a0rs*(1+zs),label=label)

print('Only valid for z <<',int(zeq(Omega_m0)))

# Case 1

Omega_L0=0.5
Omega_m0=0.5

a0rs=np.array([a0r(z) for z in zs])
label=r'$d_L\,(\Omega_{\Lambda,0},\Omega_{m,0})=($'+str(Omega_L0)+','+str(Omega_m0)+')'
plt.loglog(zs,a0rs*(1+zs),label=label)

print('Only valid for z <<',int(zeq(Omega_m0)))

# Case 1

Omega_L0=0
Omega_m0=1

a0rs=np.array([a0r(z) for z in zs])
label=r'$d_L\,(\Omega_{\Lambda,0},\Omega_{m,0})=($'+str(Omega_L0)+','+str(Omega_m0)+')'
plt.loglog(zs,a0rs*(1+zs),label=label)

print('Only valid for z <<',int(zeq(Omega_m0)))

plt.xlabel(r'$z$',fontsize=15)
plt.legend(fontsize=15)
plt.grid()
plt.tight_layout()
plt.savefig('dL.png')
plt.show()
```

```python
#compute dA and dL
plt.figure(figsize=(10,6))
zs=np.linspace(0,300,5000) # z << z_eq

# Case 2

Omega_L0=0.3
Omega_m0=0.7

a0rs=np.array([a0r(z) for z in zs])
label=r'$d_A\,(\Omega_{\Lambda,0},\Omega_{m,0})=($'+str(Omega_L0)+','+str(Omega_m0)+')'
plt.loglog(zs,a0rs/(1+zs),label=label)

# Case 1

Omega_L0=0.5
Omega_m0=0.5

a0rs=np.array([a0r(z) for z in zs])
label=r'$d_A\,(\Omega_{\Lambda,0},\Omega_{m,0})=($'+str(Omega_L0)+','+str(Omega_m0)+')'
plt.loglog(zs,a0rs/(1+zs),label=label)

# Case 1

Omega_L0=0
Omega_m0=1

a0rs=np.array([a0r(z) for z in zs])
label=r'$d_A\,(\Omega_{\Lambda,0},\Omega_{m,0})=($'+str(Omega_L0)+','+str(Omega_m0)+')'
plt.loglog(zs,a0rs/(1+zs),label=label)

plt.xlabel(r'$z$',fontsize=15)
plt.legend(fontsize=15)
plt.grid()
plt.tight_layout()
plt.savefig('dA.png')
plt.show()
```

# Question 2

```python
from scipy.special import zeta
me=5.8e9 #electron mass in kelvin
B_H = 157800 #binding energy in kelvin
func = lambda T: np.sqrt(32/np.pi)*zeta(3)*eta*(me/T)**-1.5 * np.exp(B_H/T) # (1_Xe)/X_e^2
```

```python
# Solve quadratic for Xe over temperature range and plot for different eta
eta=6e-10
Ts=np.linspace(3000,4500,1000)
Xe=np.array([np.roots([func(T),1,-1])[1] for T in Ts])
label=r'$X_{e,eq},\,\eta=$'+str(eta)
plt.plot(T,Xe,label=label)
T_est=int(T[np.abs(Xe-0.5).argmin()])
print('The temperature at which Xe = 0.5 for eta =',eta,' is approximately',T_est,'K')

eta=4e-10
Ts=np.linspace(3000,4500,1000)
Xe=np.array([np.roots([func(T),1,-1])[1] for T in Ts])
label=r'$X_{e,eq},\,\eta=$'+str(eta)
plt.plot(T,Xe,label=label)
T_est=int(T[np.abs(Xe-0.5).argmin()])
print('The temperature at which Xe = 0.5 for eta =',eta,' is approximately',T_est,'K')

eta=8e-10
Ts=np.linspace(3000,4500,1000)
Xe=np.array([np.roots([func(T),1,-1])[1] for T in Ts])
label=r'$X_{e,eq},\,\eta=$'+str(eta)
plt.plot(T,Xe,label=label)
T_est=int(T[np.abs(Xe-0.5).argmin()])
print('The temperature at which Xe = 0.5 for eta =',eta,' is approximately',T_est,'K')

plt.xlabel(r'$T\,[K]$',fontsize=15)
plt.legend(fontsize=15)
plt.savefig('Xe.png')
plt.show()
```
